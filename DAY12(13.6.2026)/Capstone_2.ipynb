```python
inventory_clean_df = inventory_df.fillna({
    "stock_quantity": 0
})

inventory_clean_df.show()
```

```python
sales_clean_df = sales_df.fillna({
    "sale_amount": 0
})

sales_clean_df.show()
```

```python
sales_clean_df = sales_clean_df.fillna({
    "payment_mode": "Not Provided"
})

sales_clean_df.show()
```

```python
products_clean_df = products_df.fillna({
    "supplier_id": "UNKNOWN"
})

products_clean_df.show()
```

```python
from pyspark.sql.functions import *

products_clean_df = products_clean_df.withColumn(
    "data_quality_status",
    when(col("supplier_id") == "UNKNOWN", "Bad")
    .otherwise("Good")
)

products_clean_df.show()
```

```python
inventory_clean_df = inventory_clean_df.withColumn(
    "data_quality_status",
    when(col("stock_quantity") == 0, "Bad")
    .otherwise("Good")
)

inventory_clean_df.show()
```

```python
sales_clean_df = sales_clean_df.withColumn(
    "data_quality_status",
    when(
        (col("sale_amount") == 0) |
        (col("payment_mode") == "Not Provided"),
        "Bad"
    ).otherwise("Good")
)

sales_clean_df.show()
```

Transformations

```python
inventory_products_df = inventory_products_df.withColumn(
    "stock_status",
    when(
        col("stock_quantity") <= col("reorder_level"),
        "Reorder Required"
    ).otherwise("Sufficient Stock")
)

inventory_products_df.show()
```

```python
products_clean_df = products_clean_df.withColumn(
    "price_category",
    when(col("unit_price") >= 50000, "Premium")
    .when(col("unit_price") >= 10000, "Standard")
    .otherwise("Budget")
)

products_clean_df.show()
```

```python
sales_clean_df = sales_clean_df.withColumn(
    "revenue_category",
    when(col("sale_amount") >= 50000, "High Revenue")
    .when(col("sale_amount") >= 15000, "Medium Revenue")
    .otherwise("Low Revenue")
)

sales_clean_df.show()
```

```python
sales_clean_df = sales_clean_df.withColumn(
    "month",
    month(col("sale_date"))
)

sales_clean_df.show()
```

```python
sales_clean_df = sales_clean_df.withColumn(
    "year",
    year(col("sale_date"))
)

sales_clean_df.show()
```

```python
inventory_products_df = inventory_products_df.withColumn(
    "inventory_value",
    col("stock_quantity") * col("unit_price")
)

inventory_products_df.show()
```

```python
suppliers_flat_df = suppliers_flat_df.withColumn(
    "supplier_quality",
    when(col("rating") >= 4.5, "Excellent")
    .when(col("rating") >= 4.0, "Good")
    .otherwise("Average")
)

suppliers_flat_df.show()
```

Window Functions

```python
from pyspark.sql.window import Window

product_window = Window.orderBy(
    col("total_revenue").desc()
)

product_revenue_df = retail_df.groupBy(
    "product_id",
    "product_name"
).agg(
    sum("sale_amount").alias("total_revenue")
)

product_revenue_df.withColumn(
    "rank",
    rank().over(product_window)
).show()
```

```python
store_window = Window.orderBy(
    col("total_revenue").desc()
)

store_revenue_df = retail_df.groupBy(
    "store_id",
    "store_name"
).agg(
    sum("sale_amount").alias("total_revenue")
)

store_revenue_df.withColumn(
    "rank",
    rank().over(store_window)
).show()
```

```python
category_window = Window.partitionBy(
    "category"
).orderBy(
    col("total_revenue").desc()
)

category_revenue_df = retail_df.groupBy(
    "category",
    "product_id",
    "product_name"
).agg(
    sum("sale_amount").alias("total_revenue")
)

category_revenue_df.withColumn(
    "rank",
    rank().over(category_window)
).show()
```

```python
category_revenue_df.withColumn(
    "rank",
    rank().over(category_window)
).filter(
    col("rank") == 1
).show()
```

```python
category_revenue_df.withColumn(
    "rank",
    rank().over(category_window)
).filter(
    col("rank") <= 3
).show()
```

```python
state_window = Window.partitionBy(
    "state"
).orderBy(
    col("total_revenue").desc()
)

state_store_df = retail_df.groupBy(
    "state",
    "store_id",
    "store_name"
).agg(
    sum("sale_amount").alias("total_revenue")
)

state_store_df.withColumn(
    "rank",
    rank().over(state_window)
).filter(
    col("rank") == 1
).show()
```

```python
running_window = Window.orderBy(
    "sale_date"
)

sales_clean_df.withColumn(
    "running_total",
    sum("sale_amount").over(running_window)
).show()
```

```python
lag_window = Window.orderBy("sale_date")

sales_clean_df.withColumn(
    "previous_sale",
    lag("sale_amount").over(lag_window)
).show()
```

```python
sales_clean_df.withColumn(
    "next_sale_amount",
    lead("sale_amount").over(lag_window)
).show()
```

```python
sales_growth_df = sales_clean_df.withColumn(
    "previous_sale",
    lag("sale_amount").over(lag_window)
)

sales_growth_df.filter(
    col("sale_amount") > col("previous_sale")
).show()
```
