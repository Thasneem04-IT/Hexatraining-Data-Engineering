```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("Working with files") \
.getOrCreate()
```

```python
from google.colab import files
files.upload()
```

```python
df = spark.read.csv(
"bookings.csv",
header=True,
inferSchema=True
)
df.show()
```

```python
df.select("passenger_name","from_city","to_city").show()
```

```python
df.filter(df.status == "Confirmed").show()
```

```python
df.filter(df.status == "Cancelled").show()
```

```python
df.filter(df.status == "Pending").show()
```

```python
df.filter(df.ticket_price > 20000).show()
```

```python
df.filter(df.to_city.isin("Dubai","Singapore","London")).show()
```

```python
df.count()
```

```python
df.groupBy("airline").count().show()
```

```python
df.groupBy("status").count().show()
```

```python
from pyspark.sql.functions import sum

df.filter(df.status == "Confirmed").select(sum("ticket_price")).show()
```

```python
from pyspark.sql.functions import avg

df.groupBy("airline").avg("ticket_price").show()
```

```python
from pyspark.sql.functions import max

df.select(max("ticket_price")).show()
```

```python
from pyspark.sql.functions import min
df.select(min("ticket_price")).show()
```

```python
df.orderBy(
    df.ticket_price.desc()
).show()
```

```python
from pyspark.sql.functions import col
df = df.withColumn("tax",col("ticket_price")*0.05)
df.show()
```

```python
df = df.withColumn("final_price",col("ticket_price")+col("tax"))
df.show()
```

```python
from pyspark.sql.functions import when
df = df.withColumn("price_category",when(col("ticket_price")>=30000,"Premium")
.when(col("ticket_price")>=10000,"Standard").otherwise("Budget"))
df.show()
```

```python
df.groupBy("price_category").count().show()
```

```python

confirmed_df = df.filter(
    df.status == "Confirmed"
)

confirmed_df.write.mode("overwrite").parquet(
    "confirmed_flights.parquet"
)
```

```python
parquet_df = spark.read.parquet(
    "confirmed_flights.parquet"
)

parquet_df.show()
```

```python
from pyspark.sql.functions import col

hotels_df = hotels_df.withColumn(
    "total_potential_revenue",
    col("rooms_available") * col("price_per_night")
)

hotels_df.show(truncate=False)
```

```python
hotels_df.orderBy(
    hotels_df.rating.desc()
).show(truncate=False)
```

```python
flattened_df = hotels_df.select(
    "hotel_id",
    "hotel_name",
    "city",
    "category",
    "rating",
    "rooms_available",
    "price_per_night",
    "amenities",
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email"),
    "total_potential_revenue"
)

flattened_df.show(truncate=False)

flattened_df.write.mode("overwrite").parquet(
    "hotels_flattened.parquet"
)
```
