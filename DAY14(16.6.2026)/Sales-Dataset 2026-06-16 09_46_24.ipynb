Window Functions

```python
window_spec = Window.orderBy(col("total_amount").desc())
```

```python
df_calc.withColumn(
    "row_number",
    row_number().over(window_spec)
).show()
```

```python
df_calc.withColumn("rank",rank().over(window_spec)).show()
```

```python
df_calc.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
).show()
```

```python
df_calc.withColumn(
    "rank",
    rank().over(window_spec)
).filter(col("rank") <= 5).show()
```

```python
df_calc.orderBy(
    col("total_amount").desc()
).limit(1).show()
```

```python
city_window = Window.partitionBy("city")\
.orderBy(col("total_amount").desc())

df_calc.withColumn(
    "rank",
    row_number().over(city_window)
).filter(col("rank")==1).show()
```

```python
cat_window = Window.partitionBy("category")\
.orderBy(col("total_amount").desc())

df_calc.withColumn(
    "rank",
    row_number().over(cat_window)
).filter(col("rank")==1).show()
```

```python
running_window = Window.orderBy("order_id")

df_calc.withColumn(
    "running_total",
    sum("total_amount").over(running_window)
).show()
```

```python
df_calc.withColumn(
    "prev_revenue",
    lag("total_amount",1).over(window_spec)
).show()
```

```python
df_calc.withColumn(
    "next_revenue",
    lead("total_amount",1).over(window_spec)
).show()
```

```python
df_calc.withColumn(
    "prev_revenue",
    lag("total_amount").over(window_spec)
).withColumn(
    "difference",
    col("total_amount") - col("prev_revenue")
).show()
```

```python
city_window = Window.partitionBy("city")\
.orderBy(col("total_amount").desc())

df_calc.withColumn(
    "rank",
    row_number().over(city_window)
).filter(col("rank") <= 2).show()
```

```python
cat_window = Window.partitionBy("category")\
.orderBy(col("total_amount").desc())

df_calc.withColumn(
    "rank",
    row_number().over(cat_window)
).filter(col("rank") <= 3).show()
```

```python
df_calc.withColumn(
    "rank",
    dense_rank().over(window_spec)
).orderBy("rank").show()
```

Mini ETL

```python
source_df = df
```

```python
transformed_df = source_df.withColumn(
    "total_amount",
    col("quantity") * col("price")
)
```

```python
completed_df = transformed_df.filter(
    col("status")=="Completed"
)
```

```python
revenue_df = completed_df.withColumn(
    "revenue",
    col("total_amount")
)
```

```python
city_report = revenue_df.groupBy("city")\
.agg(sum("revenue").alias("city_revenue"))
```

```python
category_report = revenue_df.groupBy("category")\
.agg(sum("revenue").alias("category_revenue"))
```

```python
revenue_df.write\
.mode("overwrite")\
.parquet("/FileStore/sales_parquet")
```
