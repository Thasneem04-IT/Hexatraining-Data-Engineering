Distinct and Frequency analysis

```python
print(rdd.distinct().collect())

print( rdd.distinct().count())

print(rdd.distinct().sortBy(lambda x: x).collect())

print( rdd.distinct().filter(lambda x: "PAYMENT" in x).collect())

print(rdd.distinct().filter(lambda x: "CHECK" in x).collect())

freq_rdd = rdd.map(lambda x: (x, 1)).reduceByKey(lambda a, b: a + b)

print( freq_rdd.collect())

print(freq_rdd.sortBy(lambda x: -x[1]).first())

print(freq_rdd.sortBy(lambda x: x[1]).first())

print(freq_rdd.collect())

print(freq_rdd.filter(lambda x: x[1] > 5).collect())
```

Union an Intersection

```python
extra_transactions = [
"DEPOSIT",
"ATM_WITHDRAWAL",
"TRANSFER",
"MOBILE_RECHARGE",
"BALANCE_CHECK"
]

rdd2 = spark.sparkContext.parallelize(extra_transactions)

print(rdd.union(rdd2).collect())

print(rdd.union(rdd2).count())

print(rdd.union(rdd2).distinct().collect())

print(rdd2.subtract(rdd).collect())

union_freq = rdd.union(rdd2).map(lambda x: (x,1)).reduceByKey(lambda a,b: a+b)

print(union_freq.collect())

rdd3 = spark.sparkContext.parallelize([
"DEPOSIT",
"TRANSFER",
"BALANCE_CHECK",
"CREDIT_CARD_PAYMENT",
"ATM_WITHDRAWAL"
])

print(rdd.intersection(rdd3).collect())

print(rdd.intersection(rdd3).count())

print(rdd.intersection(rdd3).sortBy(lambda x: x).collect())

print(rdd.intersection(rdd3).filter(lambda x: "CHECK" in x).collect())

print(rdd.intersection(rdd3).filter(lambda x: "TRANSFER" in x).collect())
```

Numeric RDDs

```python
amounts = [
500,1200,3000,4500,800,950,
1100,2500,4000,6000,
750,1800,2200,3300,4700,
5200,1500,2600,3700,4800,
5900,6200,700,900,1400,
1750,2100,2900,3600,4300,
5100,5800,6500,7200,8100,
9200,10500,12000,13500,15000,
17000,19000,22000,25000,28000,
30000,32000,35000,38000,40000
]

amount_rdd = spark.sparkContext.parallelize(amounts)

print(amount_rdd.sum())

print(amount_rdd.mean())

print(amount_rdd.max())

print(amount_rdd.min())

print(amount_rdd.filter(lambda x: x > 10000).collect())

print(amount_rdd.filter(lambda x: x < 2000).collect())

print(amount_rdd.map(lambda x: x + 500).collect())

print(amount_rdd.map(lambda x: x * 1.10).collect())

print(amount_rdd.sortBy(lambda x: x).collect())

print(amount_rdd.sortBy(lambda x: -x).collect())

print(amount_rdd.takeOrdered(10, key=lambda x: -x))

print(amount_rdd.takeOrdered(10))

print(amount_rdd.filter(lambda x: x > 5000).count())

print(amount_rdd.filter(lambda x: x > 20000).count())

print(amount_rdd.filter(lambda x: x > 10000).sum())
```

FlatMap

```python
customer_comments = [
"service was excellent",
"mobile banking is easy",
"transfer completed quickly",
"deposit process was smooth",
"excellent customer support",
"loan payment completed successfully",
"banking app is excellent",
"service is fast and reliable"
]

comment_rdd = spark.sparkContext.parallelize(customer_comments)

words_rdd = comment_rdd.flatMap(lambda x: x.split(" "))

print(words_rdd.collect())

print(words_rdd.count())

print(words_rdd.distinct().collect())

print(words_rdd.sortBy(lambda x: x).collect())

word_freq = words_rdd.map(lambda x: (x,1)).reduceByKey(lambda a,b: a+b)

print(word_freq.collect())

print( words_rdd.filter(lambda x: "service" in x).collect())

print(words_rdd.filter(lambda x: "excellent" in x).collect())

print(words_rdd.map(lambda x: x.upper()).collect())

print(words_rdd.max(key=len))

print(words_rdd.min(key=len))

print(word_freq.sortBy(lambda x: -x[1]).first())

print(word_freq.sortBy(lambda x: x[1]).first())

print(words_rdd.filter(lambda x: x.startswith("b")).count())

print( words_rdd.filter(lambda x: x.endswith("ly")).count())

print(word_freq.collect())
```
