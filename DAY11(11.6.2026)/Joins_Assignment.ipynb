```python
patients_df = spark.read.csv(
    "patients.csv",
    header=True,
    inferSchema=True
)

appointments_df = spark.read.csv(
    "appointments.csv",
    header=True,
    inferSchema=True
)

preferences_df = spark.read.option(
    "multiline",
    "true"
).json("patient_preferences.json")
```

```python
patients_df = patients_df.fillna({
    "city":"Unknown",
    "blood_group":"Not Available"
})

appointments_df = appointments_df.fillna({
    "consultation_fee":0
})
```

```python
final_df = patients_df.join(
    appointments_df,
    "patient_id",
    "left"
).join(
    preferences_df,
    "patient_id",
    "left"
)
```

```python
final_df = final_df.withColumn(
    "age_group",
    when(col("age") < 30,"Young")
    .when(col("age") < 50,"Adult")
    .otherwise("Senior")
)
```

```python
department_revenue = final_df.groupBy(
    "department"
).agg(
    sum("consultation_fee").alias("revenue")
)
```

```python
patient_spending = final_df.groupBy(
    "patient_id",
    "patient_name"
).agg(
    sum("consultation_fee").alias("total_spending")
)
```

```python
final_df.write.mode("overwrite") \
.parquet("hospital_analytics_output")
```
