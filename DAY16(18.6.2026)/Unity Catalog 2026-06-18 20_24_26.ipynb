# Metastore Level Storage
- Metastore is attached to data lake
- Catalog is managed
- Schema is managed
- Object is managed


```python
%sql

CREATE CATALOG IF NOT EXISTS managed_catalog;
CREATE SCHEMA IF NOT EXISTS managed_catalog.managed_schema;

CREATE TABLE IF NOT EXISTS managed_catalog.managed_schema.managed_table (id INT,name STRING);

INSERT INTO managed_catalog.managed_schema.managed_table VALUES (1, 'John'), (2, 'Mary'), (3, 'Mike');
```

```python
%sql
select * from managed_catalog.managed_schema.managed_table;
```

# Catalog is also attached to Data Lake

```python
%sql
CREATE CATALOG IF NOT EXISTS ext_catalog
MANAGED LOCATION 'abfss://metastore@atsdatabricksunity.dfs.core.windows.net/catalog';

CREATE SCHEMA IF NOT EXISTS ext_catalog.man_schema_ext_catalog;

CREATE TABLE IF NOT EXISTS ext_catalog.man_schema_ext_catalog.managed_table (id INT,name STRING);

INSERT INTO ext_catalog.man_schema_ext_catalog.managed_table VALUES (1, 'John'), (2, 'Mary'), (3, 'Mike');
```

```python
%sql
select * from ext_catalog.man_schema_ext_catalog.managed_table;
```

# Schema is also attached to the Data Lake

```python
%sql
CREATE CATALOG IF NOT EXISTS ext_catalog
MANAGED LOCATION 'abfss://metastore@atsdatabricksunity.dfs.core.windows.net/catalog';

CREATE SCHEMA IF NOT EXISTS ext_catalog.ext_schema
MANAGED LOCATION 'abfss://metastore@atsdatabricksunity.dfs.core.windows.net/schema';

CREATE TABLE IF NOT EXISTS ext_catalog.ext_schema.managed_table (id INT,name STRING);

INSERT INTO ext_catalog.ext_schema.managed_table VALUES (1, 'John'), (2, 'Mary'), (3, 'Mike');

    
```

```python
%sql
DROP TABLE ext_catalog.ext_schema.managed_table;
```

We can undrop the table within 7 days of deleting. After 7 days the data will be automatically deleted completely.

```python
%sql
UNDROP TABLE ext_catalog.ext_schema.managed_table;
```

```python
%sql
select * from ext_catalog.ext_schema.managed_table;
```

## When a table that is stored in a managed catalog is dropped, both the table metadata and the underlying data files are deleted because Databricks manages the storage location and the lifecycle of the data.

## In the case of an external we own the data and storage location. When a table is dropped, only the metadata entry is deleted from Unity Catalog, while the actual data files remain intact in the external storage (ADLS/S3/GCS).


## External Tables [Object level storage]

Object level storage..
Object is also attached to the data lake


```python
%sql
CREATE CATALOG IF NOT EXISTS ext_catalog
MANAGED LOCATION 'abfss://metastore@atsdatabricksunity.dfs.core.windows.net/catalog';

CREATE SCHEMA IF NOT EXISTS ext_catalog.ext_schema
MANAGED LOCATION 'abfss://metastore@atsdatabricksunity.dfs.core.windows.net/schema';

CREATE TABLE IF NOT EXISTS ext_catalog.ext_schema.ext_table (id INT,name STRING)
LOCATION 'abfss://raw@atsdatabricksunity.dfs.core.windows.net/ext_table';

INSERT INTO ext_catalog.ext_schema.ext_table VALUES (1, 'John'), (2, 'Mary'), (3, 'Mike');
```

```python
%sql
SELECT * FROM ext_catalog.ext_schema.ext_table;
```
