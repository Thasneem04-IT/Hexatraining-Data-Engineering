```python
journey = spark.table(
    "airline_catalog.gold.passenger_journey_gold"
)

display(journey)
```

```python
journey.write \
    .mode("overwrite") \
    .saveAsTable(
        "airline_catalog.gold.managed_table_demo"
    )
```

```python
%sql
SELECT *
FROM airline_catalog.gold.managed_table_demo;
```
