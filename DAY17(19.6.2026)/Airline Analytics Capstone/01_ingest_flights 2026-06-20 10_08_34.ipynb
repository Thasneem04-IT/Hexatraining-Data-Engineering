```python
flights_path = "/Volumes/airline_catalog/bronze/raw_files/Flights.csv"

df_flights = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(flights_path)
)

display(df_flights)
```

```python
df_flights.printSchema()
```

```python
df_flights.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("airline_catalog.bronze.flights_bronze")
```
