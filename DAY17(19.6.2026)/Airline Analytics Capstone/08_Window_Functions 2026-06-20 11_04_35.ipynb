```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
journey = spark.table(
    "airline_catalog.gold.passenger_journey_gold"
)
```

```python
flight_rev = journey.groupBy(
    "flight_id"
).agg(
    sum("revenue").alias("flight_revenue")
)
```

```python
window_spec = Window.orderBy(
    desc("flight_revenue")
)

ranked = flight_rev.withColumn(
    "rank",
    rank().over(window_spec)
)

display(ranked)
```

```python
top3 = ranked.filter(
    col("rank") <= 3
)

display(top3)
```

```python
window_spec = Window.orderBy(
    "booking_date"
)

running = journey.withColumn(
    "running_revenue",
    sum("revenue").over(window_spec)
)

display(running)
```

```python
airline_rev = journey.groupBy(
    "airline"
).agg(
    sum("revenue").alias("revenue")
)

window_spec = Window.orderBy(
    desc("revenue")
)

airline_rank = airline_rev.withColumn(
    "rank",
    rank().over(window_spec)
)

display(airline_rank)
```

```python
dest = journey.groupBy(
    "to_city"
).count()

window_spec = Window.orderBy(
    desc("count")
)

dest_rank = dest.withColumn(
    "dense_rank",
    dense_rank().over(window_spec)
)

display(dest_rank)
```
