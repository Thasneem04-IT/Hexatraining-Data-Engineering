```python
import pandas as pd
import matplotlib.pyplot as plt

journey = spark.table(
    "airline_catalog.gold.passenger_journey_gold"
)

display(journey)
```

```python
pdf = journey.toPandas()
```

Report 1: Revenue by Airline

```python
revenue_airline = pdf.groupby(
    "airline"
)["revenue"].sum()

plt.figure(figsize=(8,5))
revenue_airline.plot(kind="bar")

plt.title("Revenue by Airline")
plt.xlabel("Airline")
plt.ylabel("Revenue")

plt.show()
```

Report 2: Revenue by Travel Class

```python
revenue_class = pdf.groupby(
    "travel_class"
)["revenue"].sum()

plt.figure(figsize=(6,6))

revenue_class.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Revenue by Travel Class")

plt.show()
```

Report 3: Flights by Status

```python
status_count = pdf["status"].value_counts()

plt.figure(figsize=(6,6))

status_count.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.title("Flights by Status")

plt.show()
```

Report 4: Top Routes

```python
pdf["route"] = (
    pdf["from_city"]
    + " → "
    + pdf["to_city"]
)
```

```python
route_rev = pdf.groupby(
    "route"
)["revenue"].sum()

route_rev.sort_values().plot(
    kind="barh",
    figsize=(8,5)
)

plt.title("Top Routes")
plt.xlabel("Revenue")

plt.show()
```

Report 5: Ticket Price Distribution

```python
plt.figure(figsize=(8,5))

plt.scatter(
    pdf.index,
    pdf["ticket_price"]
)

plt.title(
    "Ticket Price Distribution"
)

plt.xlabel("Booking")
plt.ylabel("Ticket Price")

plt.show()
```
