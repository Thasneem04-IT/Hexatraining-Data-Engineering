```python
bookings_path = "/Volumes/airline_catalog/bronze/raw_files/bookings.csv - Sheet1.csv"

df_bookings = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(bookings_path)
)

display(df_bookings)
```

```python
df_bookings.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("airline_catalog.bronze.bookings_bronze")
```
