```python
preferences_path = "/Volumes/airline_catalog/bronze/raw_files/passenger_preferences.json"

df_pref = (
    spark.read
    .option("multiline", "true")
    .json(preferences_path)
)

display(df_pref)
```

```python
df_pref.printSchema()
```

```python
df_pref.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("airline_catalog.bronze.preferences_bronze")
```
