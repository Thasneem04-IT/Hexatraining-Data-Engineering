```python
from pyspark.sql.functions import *

flights = spark.table("airline_catalog.bronze.flights_bronze")

bookings = spark.table("airline_catalog.bronze.bookings_bronze")

preferences = spark.table("airline_catalog.bronze.preferences_bronze")
```

```python
display(flights)
display(bookings)
display(preferences)
```

```python
bookings = bookings.withColumn(
    "revenue",
    col("ticket_price")
)
```

```python
display(bookings)
```

```python
bookings = bookings.withColumn(
    "price_band",
    when(col("ticket_price") > 20000, "Premium")
    .when(col("ticket_price") > 10000, "Standard")
    .otherwise("Budget")
)
```

```python
display(bookings)
```

```python
flights = flights.withColumn(
    "delay_flag",
    when(col("status") == "Delayed", "Yes")
    .otherwise("No")
)
```

```python
display(flights)
```

```python
bookings.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "airline_catalog.silver.bookings_silver"
    )
```

```python
flights.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "airline_catalog.silver.flights_silver"
    )
```
