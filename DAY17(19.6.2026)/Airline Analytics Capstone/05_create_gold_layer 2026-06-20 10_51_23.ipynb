```python
flights = spark.table(
    "airline_catalog.silver.flights_silver"
)

bookings = spark.table(
    "airline_catalog.silver.bookings_silver"
)

preferences = spark.table(
    "airline_catalog.bronze.preferences_bronze"
)
```

```python
journey = bookings \
    .join(flights, "flight_id", "inner") \
    .join(preferences, "passenger_name", "left")
```

```python
display(journey)
```

```python
journey.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "airline_catalog.gold.passenger_journey_gold"
    )
```
