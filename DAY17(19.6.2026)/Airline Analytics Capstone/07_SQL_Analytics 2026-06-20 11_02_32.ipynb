```python
journey = spark.table(
    "airline_catalog.gold.passenger_journey_gold"
)

journey.createOrReplaceTempView(
    "journey"
)
```

```python
%sql
SELECT
    airline,
    SUM(revenue) AS revenue
FROM journey
GROUP BY airline
ORDER BY revenue DESC;
```

```python
%sql
SELECT
    CONCAT(from_city,' -> ',to_city) AS route,
    SUM(revenue) AS revenue
FROM journey
GROUP BY route
ORDER BY revenue DESC;
```

```python
%sql
SELECT
    AVG(ticket_price) avg_ticket_price
FROM journey;
```

```python
%sql
SELECT
    to_city,
    COUNT(*) passengers
FROM journey
GROUP BY to_city
ORDER BY passengers DESC;
```

```python
%sql
SELECT
    passenger_name,
    meal,
    seat
FROM journey;
```
