```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Orders").getOrCreate()
```

```python
data = [
(1, "Rahul Sharma", "Hyderabad", "Electronics", "Laptop", 1, 65000, "Delivered"),
(2, "Priya Reddy", "Bangalore", "Electronics", "Mobile", 2, 25000, "Delivered"),
(3, "Amit Kumar", "Mumbai", "Furniture", "Chair", 4, 3500, "Pending"),
(4, "Sneha Patel", "Chennai", "Electronics", "Tablet", 1, 30000, "Cancelled"),
(5, "Farhan Ali", "Delhi", "Furniture", "Table", 2, 12000, "Delivered"),
(6, "Neha Singh", "Hyderabad", "Fashion", "Shoes", 3, 2500, "Delivered"),
(7, "Arjun Verma", "Pune", "Fashion", "Watch", 1, 8000, "Pending"),
(8, "Meera Nair", "Bangalore", "Electronics", "TV", 1, 45000, "Delivered"),
(9, "Kiran Rao", "Hyderabad", "Furniture", "Desk", 1, 15000, "Delivered"),
(10, "Ravi Kumar", "Mumbai", "Electronics", "Mobile", 1, 28000, "Delivered")
]

columns = [
"order_id",
"customer_name",
"city",
"category",
"product",
"quantity",
"price",
"order_status"
]
df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import col
df = df.withColumn("total_amount", col("quantity")*col("price"))
df.show()
```

```python
df = df.withColumn("discount",col("total_amount")*0.10)
df.show()
```

```python
df = df.withColumn("final_amount",col("total_amount")-col("discount"))
df.show()
```

```python
df = df.withColumnRenamed("customer_name", "customer")
df.show()
```

```python
df = df.withColumnRenamed("order_status","status")
df.show()
```

```python
df = df.withColumn("order_value_category",
                   when(df.total_amount>=50000,"High").
                   when(df.total_amount>=20000, "Medium").
                   otherwise("Low"))
df.show()
```
