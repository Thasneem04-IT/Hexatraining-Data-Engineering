```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]

columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
df = spark.createDataFrame(data,columns)
df.show()
```

```python
df.filter(df.department.contains("Engineering")).show()
```

```python
df.filter(
    df.department.contains("Engineering")
).show()
```

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "salary_category",
    when(df.salary >= 85000, "High")
    .when(df.salary >= 75000, "Medium")
    .otherwise("Low")
)

df.show()
```

```python
df.groupBy(
    "salary_category"
).count().show()
```

```python
from pyspark.sql.functions import col

df = df.withColumn(
    "bonus",
    col("salary") * 0.10
)

df.show()
```

```python
df = df.withColumn(
    "annual_salary",
    col("salary") * 12
)

df.show()
```

```python
from pyspark.sql.functions import col

report_df = df.select(
    "employee_id",
    "name",
    "department",
    "salary",
    "bonus",
    "annual_salary"
)

report_df.show()
```
